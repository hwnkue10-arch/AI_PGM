```python
import matplotlib.pylab as plt
from sklearn import linear_model
# 선형 회귀 모델을 생성한다.
reg = linear_model.LinearRegression()
# 데이터는 파이썬의 리스트로 만들어도 되고 아니면 넘파이의 배열로 만들어도 됨
X = [[174], [152], [138], [128], [186]] # 학습 예제
y = [71, 55, 46, 38, 88] # 정답
reg.fit(X, y) # 학습
print("학습 완료")
print(f"기울기 : {reg.coef_}") # 기울기
print(f"절편 : {reg.intercept_}") # 절편
# 사용자 입력
height = float(input("키(cm)를 입력하세요: "))

# 예측
weight = reg.predict([[height]])

# 결과 출력
print(f"{height}cm의 예상 몸무게는 {weight[0]:.2f}kg입니다.")
# 학습 데이터를 산포도로 그린다.
plt.scatter(X, y, color='black')
# 학습 데이터를 입력으로 하여 예측값을 계산한다. 직선을 가지고 예측하기 때문에
#직선 상의 점이 된다.
y_pred = reg.predict(X)
# 예측값으로 선그래프를 그린다.
# 직선이 그려진다.
plt.plot(X, y_pred, color='blue', linewidth=4)
plt.show()
```

# 선형 회귀를 이용한 키-몸무게 예측 프로그램

## 코드 설명

이 프로그램은 `scikit-learn`의 **LinearRegression(선형 회귀)** 모델을 사용하여 키와 몸무게의 관계를 학습한 뒤, 사용자가 입력한 키를 기준으로 예상 몸무게를 예측하는 프로그램이다.

## 동작 과정

1. 필요한 라이브러리 불러오기
- `matplotlib`과 `sklearn`의 선형 회귀 기능을 가져온다.

2. 학습 데이터 준비
- `X`는 키(cm) 데이터
- `y`는 몸무게(kg) 데이터
- 입력 데이터와 정답 데이터를 이용해 모델을 학습한다.

3. 선형 회귀 모델 생성 및 학습
- `LinearRegression()`으로 모델을 생성한다.
- `fit()` 함수를 사용해 키와 몸무게의 관계를 학습한다.

4. 학습 결과 확인
- `reg.coef_` : 학습된 직선의 기울기
- `reg.intercept_` : 학습된 직선의 절편

선형 회귀 공식: 몸무게 = 기울기 × 키 + 절편

5. 사용자 입력 및 예측
- `input()`으로 사용자의 키를 입력받는다.
- `predict()` 함수를 사용해 예상 몸무게를 계산한다.

## 핵심 함수

| 함수 | 설명 |
|---|---|
| fit() | 데이터를 이용해 모델 학습 |
| predict() | 학습된 모델로 결과 예측 |
| coef_ | 모델의 기울기 확인 |
| intercept_ | 모델의 절편 확인
